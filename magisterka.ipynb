```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd

# 18.10.2021
max_page = 968
base = 'https://www.programmableweb.com/category/all/apis?deadpool=1&page='
```

```python
i = 0 
data_frame = pd.DataFrame()
while i < max_page: 
  i += 1
  url = base + str(i)
  # get the page
  print(url)
  req = requests.get(url)
  bs = BeautifulSoup(req.text, features='html.parser')
  # title 
  tds = bs.find_all('td', class_='views-field-pw-version-links')
  for td in tds: 
    innerDiv = td.find('div')
    if innerDiv == None: 
      a_tag = td.find('a')
      a_tag
      if a_tag != None:
        first_value = a_tag.text.strip().split()[0]
        data_frame = data_frame.append([first_value], ignore_index=True)
      else: 
        data_frame = data_frame.append(['N/D'], ignore_index=True)
    else:
      a_tags = innerDiv.find_all('a')
      a_tags
      a_tags_mapped = list(set(map(lambda x: x.text.strip().split()[0], a_tags)))
      first_val = list(filter(lambda x: x != 'Version', a_tags_mapped))[0]
      data_frame = data_frame.append([first_value], ignore_index=True)
    print(len(data_frame))
```
